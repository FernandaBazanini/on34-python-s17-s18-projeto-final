# Importação inicial

* As bibliotecas NumPy (np) e Pandas (pd) são importadas para facilitar a manipulação de arrays e a análise de dados em formato tabular.


```python
import numpy as np
import pandas as pd
```

# Banco de dados sobre Queimadas

* O comando pd.read_csv é utilizado para ler o arquivo CSV localizado em "/content/focos_mensal_br_202409.csv" e armazenar os dados em um DataFrame chamado dfquei.

* O comando dfquei exibe a estrutura e as primeiras linhas do DataFrame, permitindo uma visualização inicial dos dados importados.

```python
# Comando para leitura do Banco de Dados selecionado

dfquei = pd.read_csv("/content/focos_mensal_br_202409.csv")
```

```python
# Visualização inicial da base

dfquei
```

# Visualização dos Conteúdos da Base

* O método describe() é chamado no DataFrame dfquei para gerar estatísticas descritivas das colunas numéricas. Ele fornece informações como contagem, média, desvio padrão, valores mínimo e máximo, e quartis, oferecendo uma visão geral das características estatísticas dos dados.

```python
# Visualização dos conteúdos da base

dfquei.describe()
```

# Tratamento de Dados

* O comando dfquei.isnull().sum() é utilizado para verificar a existência de valores ausentes (NaN) no DataFrame dfquei. Ele retorna a soma de valores NaN em cada coluna, permitindo identificar a quantidade de dados faltantes e facilitando o tratamento dessas informações.

```python
# Verificação da existência de linhas NaN e somando-as

dfquei.isnull().sum()
```

# Exclusão de Valores NaN

* O comando dfquei.dropna() é utilizado para remover todas as linhas do DataFrame dfquei que contêm valores ausentes (NaN). Essa operação resulta em um novo DataFrame sem dados faltantes, facilitando análises posteriores com informações completas. Note que, para que a alteração tenha efeito, o resultado deve ser atribuído de volta a dfquei ou a um novo DataFrame, já que dropna() não modifica o DataFrame original por padrão

```python
# Excluindo os Nan de toda a tabela

dfquei.dropna()
```

# Banco de Dados - SP

Nesta seção, os dados do DataFrame dfquei são filtrados para criar dois novos DataFrames:

    dados_sp: Contém informações do estado de São Paulo, com as colunas estado, número de dias sem chuva, risco de fogo e precipitação.
    dados_distf: Contém dados do Distrito Federal com as mesmas colunas.

Em seguida, o método dropna() é aplicado a ambos os DataFrames para remover linhas com valores ausentes (NaN), garantindo dados completos para análises futuras.

```python
# Separando os dados de cada estado, relacionando o estado com as colunas dias sem chuva, risco de fogo e precipitação

dados_sp = dfquei[dfquei['estado'] == 'SÃO PAULO'][['estado', 'numero_dias_sem_chuva', 'risco_fogo', 'precipitacao']]
dados_distf = dfquei[dfquei['estado'] == 'DISTRITO FEDERAL'][['estado', 'numero_dias_sem_chuva', 'risco_fogo', 'precipitacao']]

# Realizando drop nas variáveis criadas para descartar possibilidade de outros NaN

dados_sp = dados_sp.dropna()
dados_distf = dados_distf.dropna()
```

# Visualizando os Dados de SP

* O método describe() é utilizado no DataFrame dados_sp para gerar estatísticas descritivas das colunas numéricas. Ele fornece informações como contagem, média, desvio padrão, valores mínimo e máximo, e quartis, oferecendo uma visão geral das características estatísticas dos dados relacionados ao estado de São Paulo.

```python
# Visualizando os dados de SP

dados_sp.describe()
```

# Removendo Números Negativos

* Este trecho identifica linhas no DataFrame dados_sp com valores negativos nas colunas risco_fogo, numero_dias_sem_chuva e precipitacao. As expressões dados_sp[(dfquei['risco_fogo'] < 0)], dados_sp[(dfquei['numero_dias_sem_chuva'] < 0)] e dados_sp[(dfquei['precipitacao'] < 0)] retornam essas linhas, mas não as removem.


```python
# Removendo os números negativos encontrado em cada coluna

dados_sp[(dfquei['risco_fogo'] < 0)]
dados_sp[(dfquei['numero_dias_sem_chuva'] < 0)]
dados_sp[(dfquei['precipitacao'] < 0)]
```

# Validando a Remoção dos Números Negativos

* Neste trecho, os números negativos são removidos do DataFrame dados_sp sem prejudicar a estrutura da tabela. O método drop() é utilizado para excluir as linhas que contêm valores negativos nas colunas risco_fogo, numero_dias_sem_chuva e precipitacao, garantindo que apenas dados válidos permaneçam no DataFrame.

```python
# Validando a remoção dos números negativos sem prejudicar a tabela

dados_sp = dados_sp.drop(dados_sp[dfquei['risco_fogo'] < 0].index)
dados_sp = dados_sp.drop(dados_sp[dfquei['numero_dias_sem_chuva'] < 0].index)
dados_sp = dados_sp.drop(dados_sp[dfquei['precipitacao'] < 0].index)
```

# Visualizando Dados de SP Após Tratamento

* O método describe() é utilizado novamente no DataFrame dados_sp para gerar estatísticas descritivas das colunas numéricas após o tratamento. Esta visualização permite verificar a contagem, média, desvio padrão, valores mínimo e máximo, e quartis, fornecendo uma visão atualizada das características estatísticas dos dados limpos e válidos relacionados ao estado de São Paulo.

```python
# Visualizando dados de SP após tratamento

dados_sp.describe()
```

# Banco de Dados - DF

No entanto, este trecho não realiza a remoção; apenas exibe os registros com valores negativos.

# Validando a Remoção dos Números Negativos sem Prejudicar a Tabela

* Neste trecho, os números negativos são removidos do DataFrame dados_distf sem afetar a estrutura da tabela. O método drop() é utilizado para excluir as linhas com valores negativos nas colunas risco_fogo, numero_dias_sem_chuva e precipitacao. Isso garante que o DataFrame dado_distf contenha apenas dados válidos e completos para análises futuras.

# Visualizando o Resultado da Remoção dos Números Negativos
